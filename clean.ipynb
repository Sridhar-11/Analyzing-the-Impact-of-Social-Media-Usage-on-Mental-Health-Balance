STEP 2 : Fix the structural issues of dataset

```python
df.head()
```


all columns are correct and no extra spaces in columns . no duplicate columns and . no mixed format

```python
# identify the null values percantege for all columns based on that wheteher to fill or drop null (<30% = fill, >30% = drop )based on column critical
percent=(df.isna().mean()/len(df)*100)
```

```python
percent
```

```python
df.describe()
```

```python
# Handle null values using mean() , median() , mode() based on the distribution of data (outliers)

df["Age"]=df["Age"].fillna(df["Age"].mean())
```

```python
df.isna().sum()
```

```python
df["Daily_Screen_Time(hrs)"]=df["Daily_Screen_Time(hrs)"].fillna(df["Daily_Screen_Time(hrs)"].mean())
```

```python
df.isna().sum()
```

```python
df["Gender"]=df["Gender"].fillna(df["Gender"].mode()[0])
```

```python
df.isna().sum()
```

```python
df["Sleep_Quality(1-10)"]=df["Sleep_Quality(1-10)"].fillna(df["Sleep_Quality(1-10)"].mean())
```

```python
df["Stress_Level(1-10)"]=df["Stress_Level(1-10)"].fillna(df["Stress_Level(1-10)"].mean())
```

```python
df["Days_Without_Social_Media"]=df["Days_Without_Social_Media"].fillna(df["Days_Without_Social_Media"].mean())
```

```python
df["Exercise_Frequency(week)"]=df["Exercise_Frequency(week)"].fillna(df["Exercise_Frequency(week)"].mean())
```

```python
df["Social_Media_Platform"]=df["Social_Media_Platform"].fillna(df["Social_Media_Platform"].mode()[0])
```

```python
df["Happiness_Index(1-10)"]=df["Happiness_Index(1-10)"].fillna(df["Happiness_Index(1-10)"].mean())
```

```python
df.isna().sum()
```

```python
# id variable dont impute blindly drop the null rows

df=df.dropna(subset="User_ID")
```

```python
df.isna().sum()
```
